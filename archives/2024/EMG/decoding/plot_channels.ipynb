```python
# Imports
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() #sets the matplotlib style to seaborn style
```

```python
data_dir = os.path.join("data", "npulse", "ch1_seq5")

# Load data per patch
for root, d_names, f_names in os.walk(data_dir):
    print('File names:', f_names)
list_data = {}
for file in f_names:
    print(f"Loading {file}...")
    data_path = os.path.join(data_dir, file)
    data = pd.read_csv(data_path)
    list_data[file] = data

# Load single data
# name = "WS_R_ch3_seq4_250523174831"
# data_path = os.path.join(data_dir, name)

# data = pd.read_csv(data_path)

print(data.head())
```

```python
# Show data
def show_channels_actions(data):
    plt.figure(figsize=(12, 6))
    for col in data.columns:
        if 'Channel' in col:
            plt.plot(data['Timestamp'], data[col], label=col, alpha=1)
            plt.legend(loc='upper left')
    
    if 'Action1' in data.columns:
        plt.twinx()
        plt.plot(data['Timestamp'], data['Action1'], label='Action 1', alpha=0.5)
        plt.legend(loc='upper right')

    plt.title('Test Data with Action')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid()
    plt.show()

def show_unique_channel(data, channel):
    plt.figure(figsize=(12, 6))
    plt.plot(data['Timestamp'], data[channel], label=channel, alpha=1)
    plt.legend(loc='upper left')
    
    if 'Action1' in data.columns:
        plt.twinx()
        plt.plot(data['Timestamp'], data['Action1'], label='Action 1', alpha=0.5)
        plt.legend(loc='upper right')

    plt.title('Test Data with Action')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid()
    plt.show()



if len(list_data) > 0:
    for key, data in list_data.items():
        show_channels_actions(data)
        # show_unique_channel(data, "Channel2")
```
